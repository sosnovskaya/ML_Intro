```python
#1
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
data = pd.read_csv('titanic.csv', index_col='PassengerId')
data[:10]
```

```python
#2
sample = data[['Survived','Pclass', 'Fare', 'Age', 'Sex']].copy()
sample[:10]
```

```python
#5
sample = sample.dropna()
sample[:10]
```

```python
#2,4
goal = sample['Survived']
goal[:10]
```

```python
features = sample[['Pclass', 'Fare', 'Age', 'Sex']]
features[:10]
```

```python
#6

def convertString(data, column):
    ndata = data.copy()
    rename = ndata[column].unique()
    renameMap = {name: n for n, name in enumerate(rename)}
    ndata[column] = ndata[column].replace(renameMap)
    return ndata

features = convertString(features, 'Sex')
features[:10]
```

```python
clf = DecisionTreeClassifier(random_state=241)
clf.fit(features, goal)
```

```python
#7
importances = pd.Series(clf.feature_importances_, index = list(features))
print(importances)
```
